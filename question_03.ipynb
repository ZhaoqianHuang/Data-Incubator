```python
industry_df = df[['industry','dataset_id']].groupby('industry').count()
industry_df.columns = ['number_industry']
plot_df = industry_df.sort_values(by = 'number_industry',ascending = False)
plt.figure(figsize = (10,10))
plt.title('Pie chart for top 15 industries from LinkedIn dataset.',fontsize=20)
plt.pie(plot_df[:15]['number_industry'],labels=plot_df[:15].index)
plt.show()
```

```python
company_df = df[['company_name','employees_on_platform']].groupby('company_name').mean()
company_df.columns = ['number_employee']
company_df_sorted = company_df.sort_values(by = 'number_employee',ascending = False)
```

```python
plt.figure(figsize = (10,10))
plt.title('Number of employees for top ten companies\n(average number from 2015-09-14 to 2018-07-17)',fontsize=20)
plt.bar(list(range(10)),company_df_sorted[:10]['number_employee'])
plt.xticks(list(range(10)), company_df_sorted[:10].index, rotation='vertical')
plt.ylabel('Number of employees',fontsize = 15)
plt.xlabel('Company name',fontsize = 15)
plt.show()
```

```python
company_followers = df[['company_name','followers_count']].groupby('company_name',as_index=False).mean()
company_employees = df[['company_name','employees_on_platform']].groupby('company_name',as_index=False).mean()
company_employees = company_employees.merge(company_followers)
```

```python
company_employees['ratio'] = company_employees['followers_count']/company_employees['employees_on_platform']
company_employees_sorted = company_employees[company_employees['employees_on_platform']>0].sort_values(by = 'ratio',ascending = False)
```

```python
plt.figure(figsize = (10,10))
plt.title('Number of employees for top ten companies\n(average number from 2015-09-14 to 2018-07-17)',fontsize=20)
plt.bar(list(range(10)),company_employees_sorted[:10]['ratio'])
plt.xticks(list(range(10)), company_employees_sorted[:10]['company_name'], rotation='vertical')
plt.ylabel('Number of employees',fontsize = 15)
plt.xlabel('Company name',fontsize = 15)
plt.show()
```

```python
IBM_df = df[df['company_name']=='IBM'][['as_of_date','followers_count','employees_on_platform']]
IBM_df['as_of_date'] = pd.to_datetime(IBM_df['as_of_date'])
IBM_df.set_index('as_of_date',inplace = True)
```

```python
IBM_stock_weekly = pd.read_csv('EOD-IBM_weekly.csv')
IBM_stock_weekly['Date'] = pd.to_datetime(IBM_stock_weekly['Date'])
IBM_stock_daily = pd.read_csv('EOD-IBM_daily.csv')
IBM_stock_daily['Date'] = pd.to_datetime(IBM_stock_daily['Date'])
IBM_stock_monthly = pd.read_csv('EOD-IBM_monthly.csv')
IBM_stock_monthly['Date'] = pd.to_datetime(IBM_stock_monthly['Date'])
```

```python
IBM_stock_daily['avg_price'] = (IBM_stock_daily['Open']+IBM_stock_daily['Close'])/2
IBM_df_daily = IBM_df.copy()
IBM_df_daily.reset_index(inplace = True)
IBM_df_daily.columns = ['Date', 'followers_count', 'employees_on_platform']
IBM_df_daily_merged = IBM_df_daily.merge(IBM_stock_daily[['Date','avg_price']],how ='left')
IBM_df_daily_merged.dropna(inplace = True)
IBM_df_daily_merged['employees_before'] = IBM_df_daily_merged['employees_on_platform'].shift(1)
IBM_df_daily_merged['employees_rate'] = (IBM_df_daily_merged['employees_on_platform']-IBM_df_daily_merged['employees_before'])/IBM_df_daily_merged['employees_before']
IBM_df_daily_merged['followers_before'] = IBM_df_daily_merged['followers_count'].shift(1)
IBM_df_daily_merged['followers_rate'] = (IBM_df_daily_merged['followers_count']-IBM_df_daily_merged['followers_before'])/IBM_df_daily_merged['followers_before']
IBM_df_daily_merged['price_before'] = IBM_df_daily_merged['avg_price'].shift(1)
IBM_df_daily_merged['price_rate'] = (IBM_df_daily_merged['avg_price'] - IBM_df_daily_merged['price_before'])/IBM_df_daily_merged['price_before']

```

```python
plt.figure(figsize = (10,10))
plt.title('The comparision of the trend between number of employees\nand stock price based on daily data(company: IBM)',fontsize=20)
plt.plot(IBM_df_daily_merged['Date'],IBM_df_daily_merged['employees_rate'])
plt.plot(IBM_df_daily_merged['Date'],IBM_df_daily_merged['price_rate'])
#plt.xticks(list(range(IBM_df_daily_merged.shape[0])), IBM_df_daily_merged['Date'], rotation='vertical')
plt.ylabel('ratios of the change of employees and stock price',fontsize = 15)
plt.xlabel('Date',fontsize = 15)
plt.legend()
plt.show()
```

```python
IBM_stock_weekly['avg_price'] = (IBM_stock_weekly['Open']+IBM_stock_weekly['Close'])/2
IBM_df_weekly = IBM_df.resample('W').mean()
IBM_df_weekly.reset_index(inplace = True)
IBM_df_weekly.columns = ['Date', 'followers_count', 'employees_on_platform']
IBM_df_weekly_merged = IBM_df_weekly.merge(IBM_stock_weekly[['Date','avg_price']],how ='left')
IBM_df_weekly_merged.dropna(inplace = True)
```

```python
plt.figure(figsize = (10,10))
plt.title('The comparision of the trend between number of employees\nand stock price based on weekly data(company: IBM)',fontsize=20)
plt.plot(IBM_df_weekly_merged['Date'],IBM_df_weekly_merged['employees_rate'])
plt.plot(IBM_df_weekly_merged['Date'],IBM_df_weekly_merged['price_rate'])
#plt.xticks(list(range(IBM_df_weekly_merged.shape[0])), IBM_df_weekly_merged['Date'], rotation='vertical')
plt.ylabel('ratios of the change of employees and stock price',fontsize = 15)
plt.xlabel('Date',fontsize = 15)
plt.legend()
plt.show()
```
